```python
import glob
import os.path

import pandas as pd
```

```python
data_dir = "../data/"
```

```python
case_files = glob.glob("*.csv", root_dir=data_dir)
```

```python
import os

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.read_csv(os.path.join(data_dir, case_files[0])).set_index(["window", "campaign_id"]).unstack(fill_value=0).cumsum()
```

```python
fig, axs = plt.subplots(len(case_files), figsize=(20, 20))

for ax, file in enumerate(case_files):
    title = os.path.basename(file)
    axs[ax].set_title(title)
    df = pd.read_csv(os.path.join(data_dir, file)).set_index(["window", "campaign_id"]).unstack(fill_value=0)
    sns.lineplot(data=df, ax=axs[ax])
```
